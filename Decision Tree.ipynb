
## To study and apply Data Preprocessing techniques on given dataset. Prepare a Social_Network_Ads dataset for training with machine learning algorithm by applying suitable data preprocessing techniques. Apply decision tree algorithm for classification. Evaluate the performance using Precision, Recall, F1 score and accuracy. Also visualize the decision tree.

##Importing Required Libraries:
pandas: Used for data manipulation and analysis.

numpy: Used for numerical computations.

train_test_split: Used to split the dataset into training and testing sets.

DecisionTreeClassifier: The Decision Tree model for classification tasks.

accuracy_score, classification_report: Metrics to evaluate model performance.

plot_tree: Used to visualize the structure of the Decision Tree.

matplotlib.pyplot: A plotting library used to visualize the Decision Tree

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

# Loading the Social Network Dataset:


```python
import pandas as pd

df = pd.read_csv('/content/Social_Network_Ads.csv')

print(df.head())
```

```python

X = df.drop(['Purchased'], axis=1)  # Features
y = df['Purchased']  # Target labels

# Perform a train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the model
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
print("Accuracy Score:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

# Visualizing the Decision Tree:

```python
plt.figure(figsize=(20,10))
plot_tree(model, feature_names=X.columns, class_names=np.unique(y).astype(str), filled=True)
plt.show()
```

#Conclusion

The Decision Tree model, after preprocessing the Social_Network_Ads dataset, showed strong classification performance. The visualization clarified feature importance and decision-making paths.
